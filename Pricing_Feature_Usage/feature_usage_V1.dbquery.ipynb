```sql
with valid_customers AS (
  SELECT DISTINCT fb.customer_id_anon
  FROM production.dwh.fact_booking fb
  LEFT JOIN production.dwh.dim_customer_ticket_reseller reseller
    ON reseller.customer_id_anon = fb.customer_id_anon
  WHERE fb.date_of_checkout >= CURRENT_DATE - INTERVAL 12 MONTH
    AND fb.status_id = 1
    AND reseller.customer_id_anon IS NULL
)

SELECT 
  fb.tour_id,
  fb.tour_option_id,
  fb.booking_id,
  fb.status_id,
  fb.gmv,
  fb.nr,
  ft.ticket_id,
  ft.ticket_category_id,
  ft.purchase_amount,
  tc.ticket_category_name
  FROM production.dwh.fact_booking fb
INNER JOIN valid_customers vc 
  ON fb.customer_id_anon = vc.customer_id_anon
Left JOIN production.dwh.fact_ticket ft 
  ON fb.booking_id = ft.booking_id and fb.tour_option_id = ft.tour_option_id 
LEFT join production.dwh.dim_ticket_category tc
  ON ft.ticket_category_id = tc.ticket_category_id
LEFT JOIN production.dwh.dim_ticket_status ts
  ON ft.ticket_status_id = ts.ticket_status_id
WHERE fb.date_of_checkout >= CURRENT_DATE - INTERVAL 12 MONTH
  AND fb.status_id = 1
```
