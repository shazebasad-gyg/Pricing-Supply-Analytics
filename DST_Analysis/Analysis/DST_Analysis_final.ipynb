```python
sql = f"""
WITH event_data AS (
  SELECT
    CASE 
      WHEN h.date_id BETWEEN DATE '{yoy_pre_start_date}' AND DATE '{yoy_pre_end_date}' THEN 'PRE_LY'
      WHEN h.date_id BETWEEN DATE '{yoy_event_week_start}' AND DATE '{yoy_post_end_date}' THEN 'POST_LY'
      WHEN h.date_id BETWEEN DATE '{event_pre_start_date}' AND DATE '{event_pre_end_date}' THEN 'PRE_CY'
      WHEN h.date_id BETWEEN DATE '{event_week_start_str}' AND DATE '{event_post_end_date}' THEN 'POST_CY'
    END AS period_window,
    dl.country_name,
    h.tour_id,
    h.supplier_id,
    date_trunc('week', h.date_id) AS week_start,
    CAST(h.is_online AS INT) AS is_online
  FROM production.supply.fact_tour_review_history h
  INNER JOIN dwh.dim_location dl ON h.location_id = dl.location_id
  WHERE (
      (h.date_id BETWEEN DATE '{event_pre_start_date}' AND DATE '{event_post_end_date}')
      OR (h.date_id BETWEEN DATE '{yoy_pre_start_date}' AND DATE '{yoy_post_end_date}')
    )
    {get_tour_filter('h')}
    {get_country_filter('dl')}
)
SELECT
  period_window,
  country_name,
  COUNT(DISTINCT week_start) AS n_weeks,
  COUNT(DISTINCT tour_id) AS total_tours,
  COUNT(DISTINCT CASE WHEN is_online = 1 THEN tour_id END) AS active_tours,
  CAST(COUNT(DISTINCT CASE WHEN is_online = 1 THEN tour_id END) AS DOUBLE) / NULLIF(COUNT(DISTINCT tour_id), 0) AS share_active_tours,
  COUNT(DISTINCT supplier_id) AS total_suppliers,
  COUNT(DISTINCT CASE WHEN is_online = 1 THEN supplier_id END) AS active_suppliers,
  CAST(COUNT(DISTINCT CASE WHEN is_online = 1 THEN supplier_id END) AS DOUBLE) / NULLIF(COUNT(DISTINCT supplier_id), 0) AS share_active_suppliers
FROM event_data
WHERE period_window IS NOT NULL
GROUP BY period_window, country_name
ORDER BY country_name, 
  CASE period_window 
    WHEN 'PRE_LY' THEN 1 
    WHEN 'POST_LY' THEN 2 
    WHEN 'PRE_CY' THEN 3 
    WHEN 'POST_CY' THEN 4 
  END
"""

print("🚀 Running optimized query...")
supply_df = spark.sql(sql).toPandas()
print(f"✨ {len(supply_df)} rows")
print(supply_df)
```

```python
sql_price_change = f"""
WITH tour_level_prices AS (
  SELECT 
    p.tour_id,
    dl.country_name,
    CASE 
      WHEN p.snapshot_date BETWEEN DATE '{yoy_pre_start_date}' AND DATE '{yoy_pre_end_date}' THEN 'PRE_LY'
      WHEN p.snapshot_date BETWEEN DATE '{yoy_event_week_start}' AND DATE '{yoy_post_end_date}' THEN 'POST_LY'
      WHEN p.snapshot_date BETWEEN DATE '{event_pre_start_date}' AND DATE '{event_pre_end_date}' THEN 'PRE_CY'
      WHEN p.snapshot_date BETWEEN DATE '{event_week_start_str}' AND DATE '{event_post_end_date}' THEN 'POST_CY'
    END AS period_window,
    CASE 
      WHEN p.snapshot_date BETWEEN DATE '{yoy_pre_start_date}' AND DATE '{yoy_post_end_date}' 
        THEN DATE '{yoy_event_week_start}'
      ELSE DATE '{event_week_start_str}'
    END AS event_start,
    p.travel_date,
    p.tour_from_price_red_per_adult,
    p.tour_from_price_black_per_adult,
    p.final_price_red_per_adult,
    p.final_price_black_per_adult,
    COALESCE(SIZE(p.availability_timeslots), 0) AS timeslot_count
  FROM production.dwh.daily_tour_price_snapshot p
  INNER JOIN production.dwh.dim_tour dt ON p.tour_id = dt.tour_id
  INNER JOIN production.dwh.dim_location dl ON dt.location_id = dl.location_id
  WHERE 
    (p.snapshot_date BETWEEN DATE '{event_pre_start_date}' AND DATE '{event_post_end_date}'
     OR p.snapshot_date BETWEEN DATE '{yoy_pre_start_date}' AND DATE '{yoy_post_end_date}')
    AND p.tour_from_price_red_per_adult > 0
    {get_tour_filter('dt')}
    {get_country_filter('dl')}
)
SELECT 
  tour_id,
  country_name,
  period_window,
  MIN(tour_from_price_red_per_adult) AS from_red_price,
  MIN(tour_from_price_black_per_adult) AS from_black_price,
  
  MIN(CASE WHEN travel_date BETWEEN event_start AND ADD_MONTHS(event_start, 3) 
           THEN final_price_red_per_adult END) AS final_red_3m,
  MIN(CASE WHEN travel_date BETWEEN event_start AND ADD_MONTHS(event_start, 6) 
           THEN final_price_red_per_adult END) AS final_red_6m,
  MIN(CASE WHEN travel_date BETWEEN event_start AND ADD_MONTHS(event_start, 12) 
           THEN final_price_red_per_adult END) AS final_red_12m,
           
  MIN(CASE WHEN travel_date BETWEEN event_start AND ADD_MONTHS(event_start, 3) 
           THEN final_price_black_per_adult END) AS final_black_3m,
  MIN(CASE WHEN travel_date BETWEEN event_start AND ADD_MONTHS(event_start, 6) 
           THEN final_price_black_per_adult END) AS final_black_6m,
  MIN(CASE WHEN travel_date BETWEEN event_start AND ADD_MONTHS(event_start, 12) 
           THEN final_price_black_per_adult END) AS final_black_12m,
           
  AVG(CASE WHEN travel_date BETWEEN event_start AND ADD_MONTHS(event_start, 3) 
           THEN timeslot_count END) AS timeslots_3m,
  AVG(CASE WHEN travel_date BETWEEN event_start AND ADD_MONTHS(event_start, 6) 
           THEN timeslot_count END) AS timeslots_6m,
  AVG(CASE WHEN travel_date BETWEEN event_start AND ADD_MONTHS(event_start, 12) 
           THEN timeslot_count END) AS timeslots_12m
           
FROM tour_level_prices
WHERE period_window IS NOT NULL
GROUP BY tour_id, country_name, period_window

"""

print("🚀 Running price change analysis...")
price_change_df = spark.sql(sql_price_change).toPandas()
print(f"✨ {len(price_change_df)} rows")
print(price_change_df)
```
